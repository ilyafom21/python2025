```python
import re

def check_car_id(car_id):
    pattern = r'^([АВЕКМНОРСТУХ])(\d{3})([АВЕКМНОРСТУХ]{2})(\d{2,3})$'
    match = re.match(pattern, car_id)
    if match:
        number, region = match.group(1) + match.group(2) + match.group(3), match.group(4)
        print(f'Номер {number} валиден. Регион: {region}.')
    else:
        print('Номер не валиден.')

check_car_id('А222ВС96')
check_car_id('АБ22ВВ193')
```

```python
import re

def remove_duplicates(text):
    return re.sub(r'\b(\w+)(?:\s+\1\b)+', r'\1', text)

some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений'
print(remove_duplicates(some_string))
```
